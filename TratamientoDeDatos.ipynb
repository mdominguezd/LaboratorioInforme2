```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Lectura de datos:

```python
Data = pd.DataFrame()

for i in range(16,21):
    Dict = 'Datos/'
    D = pd.read_excel(Dict+'LasFerias20'+str(i)+'.xlsx', skiprows = 3, skipfooter = 11)
    
    D = D[1:]

    Data = Data.append(D)
    
Data = Data.rename(columns={'Unnamed: 0':'Fecha'})
Fechas = Data.apply(lambda df : df.Fecha[:11]+'23:59' if df.Fecha[-5:-3] == '24' else df.Fecha, axis=1) # Cambio 24:00 a 23:59 para poder convertir a formato datetime

Data['Fecha'] = Fechas

# Conversion de tipos de variables
Data.Fecha = pd.to_datetime(Data.Fecha,format = '%d-%m-%Y %H:%M',errors = 'coerce')
for i in range(1,14):
    Data.iloc[:,i] = pd.to_numeric(Data.iloc[:,i], errors = 'coerce')
    
Data = Data.sort_values('Fecha')

Data['Año'] = Data.Fecha.dt.year

AgnoCuarentena = Data.apply(lambda df : True if df['Año'] == 2020 else False, axis=1)

Data['Año Cuarentena'] = AgnoCuarentena

    
Data.to_csv('DatosPeriodosTodosLosAgnos.csv')
Data.to_excel('DatosPeriodosTodosLosAgnos.xlsx')
```

# Comparación año a año parámetros por medio de Graficos de caja y bigotes

```python
for i in range(13):
    fig = plt.figure(figsize=(20,10))
    plt.title('Comparacion anual de '+str(Data.iloc[:,i+1].name))
    plt.subplot(121)
    sns.boxplot(x = Data['Año'],y = Data.iloc[:,i+1])
    plt.subplot(122)
    sns.boxplot(x = Data['Año Cuarentena'],y = Data.iloc[:,i+1])
    fig.savefig('ComparacionAnual'+str(Data.iloc[:,i+1].name)+'.png')
    if i > 1:
        plt.close()
```

```python
plt.bar(['Años antes del COVID','2020'],Data.groupby('Año Cuarentena').mean().Precipitacion)
```

```python
# Graficas de series de tiempo
for j in range(5):
    fig = plt.figure(figsize=(40,10))
    plt.suptitle('Año 20'+str(16+j), fontsize=20)
    for i in range(13):
        plt.subplot(3,5,i+1)
        np.mean(Data[Data.Fecha.dt.year == 2016+j].iloc[:,i+1])
        plt.plot(Data[Data.Fecha.dt.year == 2016+j].Fecha,Data[Data.Fecha.dt.year == 2016+j].iloc[:,i+1],lw = 0.5)
        plt.title(Data.columns[i+1])
    fig.savefig('Parametros'+str(2016+j)+'.png')
```

```python
Data2020 = Data[Data['Año Cuarentena']]
Data2020 = Data2020.drop('Año', axis=1)
Data2020 = Data2020.drop('Año Cuarentena', axis=1)
Data2020.info()
```

```python
Data2020['Dias'] = (Data2020.Fecha - Data2020.iloc[0].Fecha).dt.days
Data2020PromDia = Data2020.groupby('Dias').mean().reset_index()

sns.heatmap(Data2020PromDia.corr().iloc[0:1],center=0, cmap='seismic')
```

```python
Data2020['Cuarentena'] = Data2020.apply(lambda df : True if df.Dias>84 else False, axis=1)

for i in range(13):
    fig = plt.figure()
    sns.boxplot(x = Data2020.Cuarentena, y = Data2020.iloc[:,i+1])
    fig.savefig('ComparacionPeriodoCuarentena'+str(Data2020.iloc[:,i+1].name)+'.png')
    plt.close()
```
